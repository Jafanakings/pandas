```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
from statsmodels.stats.anova import anova_lm

# Load dataset
barium = pd.read_csv(r'C:\Users\HP\Downloads\Sample Superstore- Denormalizing using Power Query.xlsx\book_exercise\BARIUM.csv')
```

```python
# Define column names based on BARIUM.DES description
column_names = [
    'chnimp', 'bchlimp', 'befile6', 'affile6', 'afdec6', 
    'befile12', 'affile12', 'afdec12', 'chempi', 'gas',
    'rtwex', 'spr', 'sum', 'fall', 'lchnimp', 
    'lgas', 'lrtwex', 'lchempi', 't', 'feb',
    'mar', 'apr', 'may', 'jun', 'jul', 
    'aug', 'sep', 'oct', 'nov', 'dec',
    'percchn'
]

# Load CSV without headers and assign column names
barium = pd.read_csv(
    r"C:\Users\HP\Downloads\Sample Superstore- Denormalizing using Power Query.xlsx\book_exercise\BARIUM.csv",  # Update your file path here
    header=None,                    # No header row in file
    names=column_names,             # Assign our column names
    sep=',',                        # Comma-separated (adjust if needed)
    engine='python'                 # Helps with parsing quirks
)
```

```python
barium
```

```python
# (i) Model with linear time trend
model1 = sm.OLS(
    barium['lchnimp'],
    sm.add_constant(barium[['lchempi', 'lgas', 'lrtwex', 'befile6', 'affile6', 'afdec6', 't']])
).fit()

print("="*80)
print("(i) MODEL WITH LINEAR TIME TREND")
print("="*80)
print(model1.summary())
```

```python
# Identify significant variables (excluding trend)
sig_vars = [var for var in model1.pvalues.index 
            if model1.pvalues[var] < 0.05 and var != 't' and var != 'const']
print(f"\nSignificant variables (p<0.05): {', '.join(sig_vars) or 'None'}")
```

```python
# (ii) Joint significance test (excluding trend) - CORRECTED
exog_vars = ['lchempi', 'lgas', 'lrtwex', 'befile6', 'affile6', 'afdec6']
hypotheses = [(f'{var} = 0') for var in exog_vars]
ftest = model1.f_test(hypotheses)

print("\n" + "="*80)
print("(ii) JOINT SIGNIFICANCE TEST (EXCLUDING TREND)")
print("="*80)
print(f"F-statistic: {ftest.fvalue:.4f}")  # CORRECTED
print(f"P-value: {ftest.pvalue:.4f}")
print("Conclusion: Variables are jointly significant" if ftest.pvalue < 0.05 
      else "Conclusion: Variables are NOT jointly significant")
```

```python
# (iii) Add monthly dummies
monthly_dummies = ['feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

# Prepare X matrix
X = sm.add_constant(barium[['lchempi', 'lgas', 'lrtwex', 'befile6', 'affile6', 'afdec6', 't'] + monthly_dummies])
model2 = sm.OLS(barium['lchnimp'], X).fit()
```

```python
# Test joint significance of monthly dummies
ftest_dummies = model2.f_test([(f'{m} = 0') for m in monthly_dummies])

print("\n" + "="*80)
print("(iii) MODEL WITH MONTHLY DUMMIES")
print("="*80)
print(f"Seasonality test F-statistic: {ftest_dummies.fvalue:.4f}")  # CORRECTED
print(f"P-value: {ftest_dummies.pvalue:.4f}")
print("Conclusion: Significant seasonality present" if ftest_dummies.pvalue < 0.05 
      else "Conclusion: No significant seasonality")
```

```python
# Compare coefficients between models
coef_comparison = pd.DataFrame({
    'Without_Dummies': model1.params,
    'With_Dummies': model2.params[model1.params.index],
    'SE_Without': model1.bse,
    'SE_With': model2.bse[model1.params.index]
})

coef_comparison['Coef_Diff'] = coef_comparison['With_Dummies'] - coef_comparison['Without_Dummies']
coef_comparison['SE_Diff_Pct'] = (coef_comparison['SE_With'] - coef_comparison['SE_Without']) / coef_comparison['SE_Without'] * 100

print("\n" + "="*80)
print("COEFFICIENT AND STANDARD ERROR COMPARISON")
print("="*80)
print(coef_comparison.round(4))
```

```python

# List of column names extracted from PRMINWGE.DES
column_names = [
    'year', 'avgmin', 'avgwage', 'kaitz', 'avgcov', 'covt', 'mfgwage', 'prdef',
    'prepop', 'prepopf', 'prgnp', 'prunemp', 'usgnp', 't', 'post74',
    'lprunemp', 'lprgnp', 'lusgnp', 'lkaitz', 'lprun_1', 'lprepop', 'lprep_1',
    'mincov', 'lmincov', 'lavgmin'
]

# Load your minwage CSV file (without headers)
minwage = pd.read_csv(
    r"C:\Users\HP\Downloads\Sample Superstore- Denormalizing using Power Query.xlsx\book_exercise\PRMINWGE.csv",  # Update with your actual file path
    header=None,                     # No header row in the file
    names=column_names,              # Assign our custom column names
    engine='python'                  # Helps with parsing
)


minwage
```

```python
minwage.info()
```

```python
# Create logarithmic variables
import numpy as np
minwage['log_prepop'] = np.log(minwage['prepop'])
minwage['log_mincov'] = np.log(minwage['mincov'])
minwage['log_usgnp'] = np.log(minwage['usgnp'])
minwage['log_prgnp'] = np.log(minwage['prgnp'])
```

```python
# Original model (eq 10.38)
model_original = smf.ols(
    'log_prepop ~ log_mincov + log_usgnp + t', 
    data=minwage
).fit()
```

```python
print("="*80)
print("ORIGINAL MODEL (EQ 10.38)")
print("="*80)
print(model_original.summary())
```

```python
# Extended model with log(prgnp)
model_extended = smf.ols(
    'log_prepop ~ log_mincov + log_usgnp + t + log_prgnp', 
    data=minwage
).fit()
```

```python
print("\n" + "="*80)
print("EXTENDED MODEL WITH log(prgnp)")
print("="*80)
print(model_extended.summary())

# Compare minimum wage effect
minwage_effect_original = model_original.params['log_mincov']
minwage_effect_extended = model_extended.params['log_mincov']
effect_change = minwage_effect_extended - minwage_effect_original
pct_change = (effect_change / abs(minwage_effect_original)) * 100

# Interpret log(prgnp) coefficient
prgnp_coef = model_extended.params['log_prgnp']
prgnp_pval = model_extended.pvalues['log_prgnp']

print("\n" + "="*80)
print("KEY RESULTS")
print("="*80)
print(f"log(prgnp) coefficient: {prgnp_coef:.4f}")
print(f"P-value: {prgnp_pval:.4f}")
print("Significant at 5% level? ", "Yes" if prgnp_pval < 0.05 else "No")
print(f"\nInterpretation: A 1% increase in Puerto Rico's GNP is associated with a "
      f"{prgnp_coef:.4f}% change in employment, holding other factors constant.")

print(f"\nMinimum wage effect change after adding log(prgnp):")
print(f"Original coefficient: {minwage_effect_original:.4f}")
print(f"New coefficient: {minwage_effect_extended:.4f}")
print(f"Absolute change: {effect_change:.4f}")
print(f"Percentage change: {pct_change:.2f}%")
```
