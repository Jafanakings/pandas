```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
import pandas as pd

# Create a copy to keep the original df intact
df_copy = df.copy()

# Ensure 'job_posted_date' is datetime
df_copy['job_posted_date'] = pd.to_datetime(df_copy['job_posted_date'])

# Extract month name
df_copy['month'] = df_copy['job_posted_date'].dt.strftime('%B')

# List of all months in order
months_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

# Create individual DataFrames for each month with variable names
for month in months_order:
    globals()[f"{month.lower()}_df"] = df_copy[df_copy['month'] == month].drop(columns=['month'])

# Example: Access January's DataFrame as january_df
```
