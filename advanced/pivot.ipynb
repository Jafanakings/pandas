```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a copy to keep the original df intact
df_copy = df.copy()

# Ensure 'job_posted_date' is datetime
df_copy['job_posted_date'] = pd.to_datetime(df_copy['job_posted_date'])

# Extract month name
df_copy['month'] = df_copy['job_posted_date'].dt.strftime('%B')

# Find top 4 job titles by posting count
top_jobs = df_copy['job_title'].value_counts().head(4).index

# Filter for top 4 jobs
df_top_jobs = df_copy[df_copy['job_title'].isin(top_jobs)]

# Create pivot table using .size() to count rows
pivot = pd.pivot_table(
    df_top_jobs,
    index='month',
    columns='job_title',
    aggfunc='size',
    fill_value=0
)

# Optional: Order months correctly
months_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]
pivot = pivot.reindex(months_order).dropna(how='all')

# Plot line chart
pivot.plot(kind='line', marker='o', figsize=(12, 6))
plt.title('Top 4 Job Titles Trend Over Time (by Month Name)')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45)
plt.legend(title='Job Title')
plt.tight_layout()
plt.show()
```
